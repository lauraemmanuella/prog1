```python
tabuleiro = [['_', '_', '_'],['_', '_', '_'],['_', '_', '_']]
cont = 0 #define de quem é a vez

def imprimirTabuleiro():
  for lista in tabuleiro:
    linha = ''
    for item in lista:
      linha = linha + ' ' + item
    print(linha)
            
def receberEntrada():
    global cont
    imprimirTabuleiro()
    cod = 'X'
    if cont % 2 == 0:
        cod = '0'
    cont += 1
    print(f'\n{cod} é a sua vez')
    while True:
        print('Informe a posicao de sua jogada')
        linha = int(input('linha: '))
        coluna = int(input('coluna: '))
        if verificarEntrada(linha, coluna, cod):
            break
    
def verificarEntrada(linha, coluna, cod):
    if tabuleiro[linha][coluna] == '_':
        tabuleiro[linha][coluna] = cod
        return True
    return False
   
def verificarVencedor():
    cod = None
    
    # verifica as 3 linhas
    for linha in tabuleiro:
        cod = verificar(linha)
        if cod != None:
            return cod
   
    # verifica as 3 colunas
    for coluna in range(3):
        cod = verificar([x[coluna] for x in tabuleiro])
        if cod != None:
            return cod

    #verifica diagonal principal
    diagonalP = [tabuleiro[0][0], tabuleiro[1][1], tabuleiro[2][2]]
    cod = verificar(diagonalP)
    if cod != None:
        return cod
    
    #verifica diagonal secundaria
    diagonalS = [tabuleiro[0][2], tabuleiro[1][1], tabuleiro[2][0]]
    cod = verificar(diagonalS)
    if cod != None:
        return cod
    
    if cont == 9:
        cod = 'VELHA'
    
    return cod

def verificar(sequencia):
    cod = None
    if sequencia.count('X') == 3:
            cod = 'X'
    elif sequencia.count('0') == 3:
            cod = '0'
    return cod
        
def main():
    print('JOGO DA VELHA')
    cod = None
    while True:
        cod = verificarVencedor()
        if cod != None:
            break
        receberEntrada()
    imprimirTabuleiro()
    print(f'\n{cod} é o vencedor!')

#executar   
main()
```
