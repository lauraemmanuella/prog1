**Questão 4: Progressão aritmética.** Escreva um programa que mostre uma progressão aritmética. Para isso, receba do usuário o primeiro termo da progressão, a razão da progressão e a quantidade de termos. 

```python
primeiro = int(input('Informe o primeiro termo da progressão artimética: '))
razao = int(input('Informe a razão da progressão artimética: '))
qtd = int(input('Informe a quantidade de termos da progressão artimética: '))
i = 1
termo = primeiro
while i <= qtd:
  print(termo)
  termo = termo + razao
  i += 1
```
