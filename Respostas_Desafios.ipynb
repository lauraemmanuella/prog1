**Questão 4:** O código abaixo implementa uma versão simples do jogo da forca. Observe o código e tente realizar os desafios 4A e 4B abaixo:

```python
import random

palavras = ['TERMO', 'VIOLA', 'AREIA', 'SOLAR', 'LIVRO', 'RATOS', 'DORME', 'DISCO', 'AJUDA', 'TEXTO']

secreta = random.choice(palavras)

print(secreta)

resposta = '_ _ _ _ _' #Questão 4A AQUI

entradas = []
acertos = []

while resposta != secreta:
  letra = input('\nDigite uma letra: ').upper()
  if letra in entradas:
    print('\nletra já usada')
  else:
    entradas.append(letra)
    if letra in secreta:
      acertos.append(letra)
      resposta = ''
      for i in secreta:
        if i in acertos:
          resposta = resposta + i
        else:
          resposta = resposta + ' _ '
  print(resposta)
  #Questão 4B AQUI
```

**Questão 4B**: Após mostrar a string resposta na última linha, mostre as entradas já colocadas, exceto os acertos, como na execução abaixo:

```python
import random

palavras = ['TERMO', 'VIOLA', 'AREIA', 'SOLAR', 'LIVRO', 'RATOS', 'DORME', 'DISCO', 'AJUDA', 'TEXTO']

secreta = random.choice(palavras)

print(secreta)

#Altere a linha 9 para que a string resposta gere os espaços (underscore) de acordo com o tamanho da string secreta
resposta = ''
for i in secreta:
  resposta = resposta + ' _ '

entradas = []
acertos = []

while resposta != secreta:
  letra = input('\nDigite uma letra: ').upper()
  if letra in entradas:
    print('\nletra já usada')
  else:
    entradas.append(letra)
    if letra in secreta:
      acertos.append(letra)
      resposta = ''
      for i in secreta:
        if i in acertos:
          resposta = resposta + i
        else:
          resposta = resposta + ' _ '
  print(resposta)
  #Mostre as entradas já colocadas, exceto os acertos
  out = ''
  for i in entradas:
    if i not in acertos:
      out = out + i + ' '
  print(f'\nLetras usadas: {out}')
```

**Questão 5:** Implemente uma versão simples do jogo termo. Observe o exemplo de execução:

```python
import random

palavras = ['TERMO', 'VIOLA', 'AREIA', 'SOLAR', 'LIVRO', 'RATOS', 'DORME', 'DISCO', 'AJUDA', 'TEXTO', 'NAVIO']

secreta = random.choice(palavras)

entrada = ''

while entrada != secreta:
  entrada = input('\nEntrada: ').upper()
  if len(entrada) == 5:
    for i in range(len(secreta)):
      if entrada[i] in secreta:
        if entrada[i] == secreta[i]:
          print(f'{entrada[i]} na posição correta')
        else:
          print(f'{entrada[i]} em posição incorreta')
      else:
        print('-')
  
print('\n Uhuuuuu')
```
